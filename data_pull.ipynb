#### Importing libraries

```python
import pandas as pd
from bs4 import BeautifulSoup as bs 
```

#### Saving the webpage that contains the files needed

wget -O webpage.html https://archive.org/serve/third-eye

#### Parsing the webpage to pull the file paths wanted

```python
filehandle = open('webpage.html')
soup = bs(filehandle, 'html.parser')

list_of_filenames = []
list_of_files = []

rows = soup.find_all('tr')

for row in rows:
    cells = row.find_all(['td', 'th'])
    
    cell_texts = [cell.get_text(strip=True) for cell in cells]
    if '.tsv' in cell_texts[0] and '-tweets.tsv' not in cell_texts[0] and '2017-' not in cell_texts[0]:
        list_of_files.append(cell_texts[0])
        list_of_filenames.append("https://archive.org/serve/third-eye/"+cell_texts[0])
```

#### Writing the list of files to pull from the site

```python
with open("list_of_files.txt", 'w') as outfile:
    for item in list_of_filenames:
        outfile.write(str(item) + '\n')
```

#### Using wget in terminal to pull the files

wget  --input-file=list_of_filenames.txt

#### After grouping each file by year into folders, the following prompts were used in terminal

##### The code below combines all of the days in the folder to have one file for all of the year's data, filters for instances where the chyron duration is greater than 1, then saves into a new file
