```python
import pandas as pd
```

```python
df_2018 = pd.read_csv("2018_clean.csv")
df_2019 = pd.read_csv("2019_clean.csv")
df_2020 = pd.read_csv("2020_clean.csv")
df_2021 = pd.read_csv("2021_clean.csv")
df_2022 = pd.read_csv("2022_clean.csv")
df_2023 = pd.read_csv("2023_clean.csv")
df_2024 = pd.read_csv("2024_clean.csv")
df_2025 = pd.read_csv("2025_clean.csv")
df_2026 = pd.read_csv("2026_clean.csv")
```

```python
df = pd.concat([df_2018, 
                df_2019,
                df_2020,
                df_2021,
                df_2022,
                df_2023,
                df_2024,
                df_2025,
                df_2026], 
                ignore_index=True) 
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df
```

```python
df = df.drop(columns=["Unnamed: 0", "text_cleaned"])
```

```python
df
```

```python
df['show_tag'] = df['tag'].str.split("/start").str[0].str.split("_").str[3:].str.join(' ')
```

```python
df = df.drop(columns=["tag", "original_chyron"])
```

```python
df['date'] = df['date'].str.replace('-', '', regex=False)

df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].dt.year
```

```python
df
```

```python
df.to_csv("combined_years.csv")
```
